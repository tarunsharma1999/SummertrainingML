***create a numpy array of  8x2 as having number  in each cell between 100 and 200 such that difference between each element is 5  ***

```
import numpy as np
```

```
array=np.arange(100,125,1)
for i in array:
  j=i+80
  result=np.arange(i,j,5).reshape(8,2)
  print(result)
```
